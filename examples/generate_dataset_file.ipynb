```python
import json
import os

from PIL import Image

with open('../../vos-visualization-tool/assets/dataset/dataset.json', 'r') as fp:
    dataset = json.load(fp)
```

```python
DAVIS_ROOT = '../../../datasets/davis-2017/data/DAVIS'

for seq in dataset['sequences'].keys():
    annotations_file = os.path.join(DAVIS_ROOT, 'Annotations', '480p', seq, '00000.png')
    img = Image.open(annotations_file)
    dataset['sequences'][seq]['image_size'] = list(img.size)
    print(img.size)
    
with open('../../davis-interactive/davisinteractive/dataset/davis.json', 'w') as fp:
    json.dump(dataset, fp)
```

```python
with open('../../davis-interactive/davisinteractive/dataset/davis.json', 'r') as fp:
    dataset = json.load(fp)
    
for k in dataset['sequences'].keys():
    dataset['sequences'][k]['num_scribbles'] = 3

with open('../../davis-interactive/davisinteractive/dataset/davis.json', 'w') as fp:
    json.dump(dataset, fp)
```
